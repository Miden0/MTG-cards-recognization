```python
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import os
from threading import Thread
from queue import Queue
```

```python
os.chdir("original")
```

```python
def getpage(page_n, queue):
    pointeur = page_n * 30
    response = requests.get(f"https://www.mtgpics.com/results?type_sel[AR]=1&type_sel[CR]=1&type_sel[EN]=1&type_sel[IN]=1&type_sel[SO]=1&type_sel[TR]=1&type_sel[PL]=1&type_sel[LA]=1&type_sel[TK]=1&type_search=1&costx=1&endx=1&forx=1&graph_aff=1&pointeur={pointeur}")
    soup = BeautifulSoup(response.content, "html.parser")
    queue.put(soup)

def saveimg(soup, img_index):
    title = soup.select(f"#r_main{img_index} .Card16")[0].text
    img_url = f'http://www.mtgpics.com{soup.select(f"#r_main{img_index} a img")[0].attrs["src"]}'
    img_url = img_url.replace("reg", "big")

    img_content = requests.get(img_url).content

    with open(f"{title}.jpg", "wb") as image:
        image.write(img_content)
```

```python
#working solution (get all pages than their images)

total_pages = 4
pages = Queue()
threads = [None, None, None, None]

for p in range(3,total_pages+1, 4):
    for i in range(4):
        page = p-i
        t = Thread(target=getpage, args=(page, pages))
        t.start()
        threads[i-1] = t
    for t in threads:
        t.join()

threads = [None, None, None, None, None]

while not pages.empty():
    page = pages.get()
    for c in range(5, 31, 5):
        for i in range(5):
            card_index = c-i
            t = Thread(target=saveimg, args=(page, card_index))
            t.start()
            threads[i-1] = t
    for t in threads:
        t.join()

print("ending")
```
